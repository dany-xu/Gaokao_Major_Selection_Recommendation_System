## 专业赋分策略

2. 进入循环，以每所大学、每一行为索引：
* 查看大学有无eva，如果没有，根据985双一流情况赋分
    有，设置平均分mean
* 对该行c2code进行迭代查找，有的话+=evaScore，没有的话+=mean，取平均
