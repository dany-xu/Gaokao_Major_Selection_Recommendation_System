```python
code=[]
c2=[]
whole_eva=[]

for addurl in urls:
    url='http://www.cdgdc.edu.cn/webrms/pages/Ranking/'+addurl
    r = requests.get(url, timeout=30)
    soup = BeautifulSoup(r.text, "html.parser")
    trs=soup.find_all('tr')
    tds=trs[0]('td')
    
    nc2=tds[3]('strong')[0].string[:4]
    code.append(nc2)
    c2.append(tds[3]('strong')[0].string[6:])
    
    for i in range(len(tds[3]('div')[0]('td'))):
        td=tds[3]('div')[0]('td')[i]
        if len(td.string)<=4:
            eva=td.string[:-1]
        else:
            college=td.string[1:6]
            new=[college,nc2,eva]
            whole_eva.append(new)
whole_eva
```

```python
import pandas as pd
a={'c2name':c2,'mjcode':code}
a=pd.DataFrame(a)
a.to_csv('c2name_mjcode.csv',encoding='gbk')
```

```python
with open('evaluation.csv', 'w', newline='')as f:#打开文件
    fw=csv.writer(f)#写入文件
    fw.writerows(whole_eva)
f.close()
```
